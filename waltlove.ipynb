```python
money_way = os.path.join("Disney Datasets","disney_revenue_1991-2016.csv")
money_df = pd.read_csv(money_way, encoding="utf-8")
money_df
```

```python
pd.isnull(money_df).sum()
```

```python
money_df = money_df.drop(['Disney Interactive[NI 3][Rev 1]'], axis=1)
money_df = money_df.dropna()
money_df
```

```python
total_revenue, = plt.plot(money_df["Year"], money_df["Total"], color="blue", label="Total Revenue" )

# Plot the unemployment values for a single country

media_revenue, = plt.plot(money_df["Year"], money_df["Disney Media Networks"].astype(float), color="green", label="Media Revenue")
studio_revenue, = plt.plot(money_df["Year"], money_df["Studio Entertainment[NI 1]"].astype(float), color="red", label="StudioENT Revenue")
product_revenue, = plt.plot(money_df["Year"], money_df["Disney Consumer Products[NI 2]"].astype(float), color="black", label ="Consumer Products")
park_revenue, = plt.plot(money_df["Year"], money_df["Walt Disney Parks and Resorts"].astype(float), color="yellow", label ="Park Revenue")


# Create a legend for our chart, color="green", label="Media Revenue")
plt.legend(handles=[total_revenue, media_revenue], loc="best")

# Show the chart
plt.show()
```
